```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('milk_production.csv')
df
```

```python
#1. Which state has produced max milk in year 2013-14?
df["Maximum_milk"]=df['Cow Milk-2013-14']+df['Boffalo Milk-2013-14']+df['Goat Milk-2013-14']
df[df['Maximum_milk']==df['Maximum_milk'].max()][['State/ UT Name','Maximum_milk']]
```

```python
#2. List top 5 milk producing states in each year from 2010 to 2015
#year 2010-11
df["Y2011"] =(df["Cow Milk-2010-11"] + df["Boffalo Milk-2010-11"] + df["Goat Milk-2010-11"]).sort_values(ascending=False)
df.sort_values('Y2011',ascending=False)["State/ UT Name"].head()
```

```python
#year 2011-12
df["Y2012"] =(df["Cow Milk-2011-12"] + df["Boffalo Milk-2011-12"] + df["Goat Milk-2011-12"]).sort_values(ascending=False)
df.sort_values('Y2012',ascending=False)["State/ UT Name"].head()
```

```python
#year 2013-14
df["Y2014"] =(df["Cow Milk-2013-14"] + df["Boffalo Milk-2013-14"] + df["Goat Milk-2013-14"]).sort_values(ascending=False)
df.sort_values('Y2014',ascending=False)["State/ UT Name"].head()
```

```python
#year 2014-15
df["Y2015"] =(df["Cow Milk-2014-15"] + df["Boffalo Milk-2014-15"] + df["Goat Milk-2014-15"]).sort_values(ascending=False)
df.sort_values('Y2015',ascending=False)["State/ UT Name"].head()
```

```python
#3. Calculate average milk production by cow, buffalo and goat in each year for selected state.

df[df['Cow Milk-2010-11'].mean() and df["State/ UT Name"]=="Karnataka"]
```

```python
#cow
df["cow_milk_allyear"]=df["Cow Milk-2010-11"]+df["Cow Milk-2011-12"]+df["Cow Milk-2013-14"]+df["Cow Milk-2014-15"]+df["Cow Milk-2015-16"]
cow=df[df["cow_milk_allyear"].mean() and df["State/ UT Name"]=="Karnataka"]
cow
```

```python
#boffalo
df["boffalo_milk_allyear"]=df["Boffalo Milk-2010-11"]+df["Boffalo Milk-2011-12"]+df["Boffalo Milk-2013-14"]+df["Boffalo Milk-2014-15"]+df["Boffalo Milk-2015-16"]
boffalo=df[df["boffalo_milk_allyear"].mean() and df["State/ UT Name"]=="Karnataka"]
boffalo
```

```python
#goat
df["goat_milk_allyear"]=df["Goat Milk-2010-11"]+df["Goat Milk-2011-12"]+df["Goat Milk-2013-14"]+df["Goat Milk-2014-15"]+df["Goat Milk-2015-16"]
goat=df[df["goat_milk_allyear"].mean() and df["State/ UT Name"]=="Karnataka"]
goat
```

```python
#4. List 5 states (if present) whose total milk production is increased for last three years.
```

```python
#year 2015-16
df["Y2016"] =(df["Cow Milk-2015-16"] + df["Boffalo Milk-2015-16"] + df["Goat Milk-2015-16"]).sort_values(ascending=False)
df.sort_values('Y2016',ascending=False)["State/ UT Name"].head()
```

```python
df[df["Y2016"].ge(df["Y2015"]) & df["Y2015"].ge(df["Y2014"])].head()
```

```python
df1=df[df["State/ UT Name"]=="Karnataka"] 
df1
```

```python
#Draw line graph showing total milk production for last 5 years of selected state




plt.plot(df1['Y2011'], df1["State/ UT Name"], linestyle="dotted", marker='^', color="red", label = 'Y2011')

plt.plot(df1['Y2012'], df1["State/ UT Name"], linestyle="solid", marker='o', color="green", label = 'Y2012')

plt.plot(df1['Y2014'], df1["State/ UT Name"], linestyle="dashed", marker='<', color="blue", label = 'Y2014')

plt.plot(df1['Y2015'], df1["State/ UT Name"], linestyle="dashed", marker='<', color="blue", label = 'Y2015')

plt.plot(df1['Y2016'], df1["State/ UT Name"], linestyle="dashed", marker='<', color="blue", label = 'Y2016')


plt.xlabel("Total Milk Produced")
plt.ylabel("State")


plt.legend()


plt.title("Milk Production in Karnataka")


plt.show()
```
